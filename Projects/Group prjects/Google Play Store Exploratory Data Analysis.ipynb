# Data Visualization

### Plot a Graph

```python
d.dropna(inplace=True)
```

```python
d
```

```python
d1=d.head(1000)
```

```python
d1
```

# Bar Graph between Category Vs Installs

```python
plt.figure(figsize=(25,10))
plt.bar(d1['Category'],d1['Installs'])
plt.title('Category Vs Installs',fontsize=15,color='red')
plt.xlabel('Category',fontsize=15)
plt.ylabel('Installs',fontsize=15)
plt.show()
```

# Bar Graph between category and content rating

```python
plt.figure(figsize=(25,10))
plt.bar('Category','Content Rating', data=d1,color='green')
plt.show()
```

# value count of each category using countplot

```python
print(d1['Category'].value_counts())
plt.figure(figsize=(15,4))
sns.countplot(x=d1['Category'])
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(10,5))
plt.bar(d1['Category'],d1['Rating'],color="red")
plt.title('Category Vs Rating',fontsize=15,color='blue')
plt.xlabel('Category',fontsize=15)
plt.ylabel('Rating',fontsize=15)
plt.show()
```

# Relplot 

```python
plt.figure(figsize=(25,5))
sns.relplot(x='Rating',y='Category',hue='Content Rating',data=d1)
plt.show()
```

```python
plt.figure(figsize=(10,7))
sns.countplot(x='Content Rating',data=d1,palette='coolwarm')
plt.show()
```

# Piechart

```python
plt.figure(figsize=(8,5))
plt.title('Content Rating',fontsize=20)
d1['Content Rating'].value_counts().plot.pie(autopct='%1.1f%%',shadow=True)
plt.show()
```

```python
plt.figure(figsize=(25,5))
sns.countplot(x='Category',data=d1,palette='tab10')
plt.show()
```

```python
d.groupby('Category')[['Rating']].sum().plot.bar(color=['red','blue'],figsize=(8,5))
plt.ylabel('Rating')
plt.show()
```

```python
d2=d1.sample(50)
plt.figure(figsize=(12,8))
plt.title('Rating',fontsize=20)
d2['Rating'].value_counts().plot.pie(autopct='%1.1f%%',shadow=True)
plt.show()
```

```python
plt.figure(figsize=(15,10))
plt.title('Category wise Rating for each App')
sns.barplot(x='Rating',y='Category',data=d1.sample(50),hue='Content Rating',palette='coolwarm_r')
plt.xlabel('Rating',fontsize=15)

plt.show()
```

```python
plt.figure(figsize=(25,8))
plt.title('Category wise Rating in each App')
sns.barplot(y='Rating',x='Category',data=d1,palette='coolwarm_r')
plt.xlabel('Category',fontsize=15)

plt.show()
```
